```python
img = Image.open("smerfy.jpg")

width, height = img.width, img.height

r0,g0,b0 = (44,184,233) # kolor niebieski - należy go odczytać ręcznie z obrazka
tol = 60 # tolerancja - dobrać eksperymentalnie

# Przejście po pikselach obrazu
for x in range(width):
    for y in range(height):
        # Odczytanie wartości piksela
        r,g,b = img.getpixel( (x,y) )
        
        # Sprawdzenie czy dany piksel ma wartość RGB koloru niebieskiego, z ustaloną tolerancją
        if abs(r-r0) < tol and abs(g-g0) < tol and abs(b-b0) < tol:
            # Zamiana piksela na kolor różowy
            img.putpixel( (x,y), ImageColor.getrgb("pink") )

# Obraz jako "wykres"
import matplotlib.pyplot as plt
plt.imshow(img)
plt.grid(False)
plt.show()

```
