```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("pokemon.csv")
df.head()
```

```python
X = df.iloc[:, 5:]
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
X_scaled = scaler.transform(X)
```

```python
pd.DataFrame(X_scaled).head()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(X_scaled)
```

```python
X_pca = pca.transform(X_scaled)
X_pca
```

```python
c = pd.factorize(df["Type 1"])
```

```python
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=c[0])
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Pokemon PCA")
plt.show()
```
