```python
import matplotlib.pyplot as plt
```

```python
def draw_line_plot(probabilities):
    plt.plot(range(11), probabilities, 'o-')
    plt.xticks(range(11))
    plt.yticks([i/10 for i in range(11)])
    plt.grid(True)
    plt.xlabel("Ilość testów pozytywnych z rzędu")
    plt.ylabel("Prawdopodobieństwo bycia chorym")
    plt.title("Zaleznosc prawdopodobienstwa bycia chorym od ilosci przeprowadzonych testow", weight="bold", fontsize=10, pad=20)
    plt.plot(range(-1, 12), [0.5]*13, 'r-')
    plt.xlim((-0.5, 10.5))
    plt.ylim((.0, 1.05))
    plt.show()
```

```python
p_ch = .1  # apriori
p_poz_given_ch = .9
p_poz_given_not_ch = .3

probabilities = [p_ch]
for i in range(10):
    p_ch_given_poz = (p_poz_given_ch*p_ch) / (p_poz_given_ch*p_ch + p_poz_given_not_ch*(1-p_ch))  # bayes theorem
    probabilities.append(p_ch_given_poz)  # collecting aposteriori probabs
    p_ch = p_ch_given_poz  # aposteriori becomes new apriori in the next inference

draw_line_plot(probabilities)
```
